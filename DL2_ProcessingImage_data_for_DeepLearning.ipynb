#Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/puppy.jpg?fmt=webp&w=500')
img_resized= img.resize((100,100))
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
img_res = mpimg.imread('/content/dog_img_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

#**Converting rgb images to Grayscale image using OpenCv**

```python
#importing OpenCV library
import cv2
```

```python
img=cv2.imread('/content/puppy.jpg?fmt=webp&w=500')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
#displaying the image
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale image
cv2.imwrite('dog_grayscale_img.jpg',grayscale_image)
```
